```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)

clientes_df = pd.read_csv('../dataset/base_clientes.csv')
churn_df = pd.read_csv('../dataset/base_hist_churn.csv')
precos_df = pd.read_csv('../dataset/base_precos.csv')

display(clientes_df)
```

```python
precos_df = precos_df.dropna()
precos_df = precos_df.drop('price_date', axis=1)

# Calcular a mdia das colunas de preos agrupadas somente por 'id'
precos_df = precos_df.groupby('id').mean().reset_index()
precos_df.info()

# Exibir o DataFrame resultante
clientes_df = pd.merge(clientes_df, precos_df, on='id')
clientes_df = pd.merge(clientes_df, churn_df, on='id')
```

```python
colunas_drop = ['activity_new', 'campaign_disc_ele', 'date_first_activ', 'forecast_base_bill_ele', 'forecast_base_bill_year', 'forecast_bill_12m', 'forecast_cons']
drop_rows = ['date_end', 'date_modif_prod', 'date_renewal', 'forecast_discount_energy', 'forecast_price_energy_p1', 'forecast_price_energy_p2', 'forecast_price_pow_p1', 'margin_gross_pow_ele', 'margin_net_pow_ele', 'net_margin', 'origin_up', 'pow_max']

clientes_df = clientes_df.drop(colunas_drop, axis=1)
clientes_df = clientes_df.dropna(subset=drop_rows)
```

```python
clientes_df = pd.get_dummies(clientes_df, columns=['channel_sales'], prefix='channel_sales')
clientes_df.isnull().sum()
```

```python
clientes_df['has_gas'] = clientes_df['has_gas'].map({'t': 1, 'f': 0})

colunas = ['cons_12m', 'cons_gas_12m', 'cons_last_month', 'forecast_cons_12m', 'forecast_cons_year', 'forecast_discount_energy', 'forecast_meter_rent_12m', 'forecast_price_energy_p1', 'forecast_price_energy_p2', 'forecast_price_pow_p1', 'has_gas', 'imp_cons', 'margin_gross_pow_ele', 'margin_net_pow_ele', 'nb_prod_act', 'net_margin', 'num_years_antig', 'pow_max', 'price_p1_var', 'price_p2_var', 'price_p3_var', 'price_p1_fix', 'price_p2_fix', 'price_p3_fix']
clientes_df = clientes_df[~(clientes_df[colunas] < 0).any(axis=1)]
clientes_df.info()
```
